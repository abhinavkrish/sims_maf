```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
%matplotlib inline
import lsst.sims.maf.sliceMetrics as sliceMetrics
import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
```

```python
filenames = !ls $outDir/*Galaxy*npz
print filenames
```

```python
sm = sliceMetrics.ComparisonSliceMetric(useResultsDb=False, outDir=outDir)
rm = sliceMetrics.RunSliceMetric(useResultsDb=False, outDir=outDir)
```

```python
iids = sm.readMetricData(filenames)
iids2 = rm.readMetricData(filenames)
for iid in iids:
    meanN = np.mean(sm.metricValues[iid].compressed())
    sm.metricValues[iid] = (sm.metricValues[iid]-meanN)/meanN
    rm.metricValues[iid] = (rm.metricValues[iid]-meanN)/meanN
    if sm.metadatas[iid].split()[-1] == 'NoDither':
        continue
    else:
        cutlevel = np.percentile(sm.metricValues[iid], 4.)
        mask = np.where(sm.metricValues[iid]<cutlevel)[0]
        sm.metricValues[iid][mask] = hp.UNSEEN
        rm.metricValues[iid][mask] = hp.UNSEEN
```

```python
rm.slicer = rm.slicers[0]
for iid in iids:
    rm.slicer.plotSkyMap(rm.metricValues[iid], title=rm.metadatas[iid], colorMin=-.4, colorMax=0.3, xlabel="dN/<N>")
```

```python
totalPower = {}
for iid in iids:
    totalPower[iid] = rm.computeSummaryStatistics(iid, metrics.TotalPowerMetric('metricdata'))[0]
iidsx = np.argsort(totalPower.values())
iidsx = iidsx
print iidsx
for iid in iidsx:
    print rm.metadatas[iid].split()[-1], sm.metadatas[iid], totalPower[iid]
```

```python
labels = []
for iid in iidsx:
    labels.append(sm.metadatas[iid].split()[-1] + ' (power) %.4g' %(totalPower[iid]))
    #print iid, labels[iid]
sm.plotPowerSpectra(iidsx, labels=labels, removeDipole=False)
plt.xlim(0, 500)
plt.yscale('linear')
plt.ylim(0, 0.0005)
plt.legend(loc=[0.9, 0.5], fontsize='smaller', fancybox=True)
```

Just checking the galaxy numbers function. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
def _galCount(coaddm5, apparent_mag):
        dn_gal = np.power(10., -3.52) * np.power(10., 0.34*apparent_mag)
        completeness = 0.5*scipy.special.erfc(apparent_mag-coaddm5)
        return dn_gal*completeness

m = np.arange(20, 30, 0.25)
coaddm5 = 27.5
y = _galCount(coaddm5, m)
constant = 41253/np.sum(y)
y = y*constant
plt.plot(m, y)
plt.xlabel('Magnitude')
plt.ylabel('NGalaxies')
```

The last few plots are investigating the functional shape of the galaxies completeness function.

```python
import scipy 
x = np.arange(24, 29.5, 0.1)
maxCompleteness = 1
m5 = 27
rolloverSpeed = 2.0
y = 0.5*(scipy.special.erfc((x-m5)/(rolloverSpeed/2.0)))
plt.plot(x, y)
plt.xlabel('Magnitude')
plt.ylabel('erfc completeness')
```

```python
x = np.arange(24, 29.5, 0.1)
maxCompleteness = 1
m5 = 27
rolloverSpeed = 2.5
y = 0.5*(maxCompleteness - np.tanh((x-m5)/(rolloverSpeed/4.0)))
plt.plot(x, y)
plt.xlabel('Magnitude')
plt.ylabel('tanh completeness')
```
